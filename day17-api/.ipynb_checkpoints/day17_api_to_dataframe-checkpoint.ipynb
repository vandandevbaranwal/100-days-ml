```python
import pandas as pd
import requests
df_list = []
for i in range(1, 21):
    response = requests.get('https://fakestoreapi.com/products/{}'.format(i))
    if response.status_code == 200 and response.json():
        temp_df = pd.DataFrame([response.json()])[['id', 'title']]
        df_list.append(temp_df)

if df_list:
    df = pd.concat(df_list, ignore_index=True)
else:
    df = pd.DataFrame(columns=['id', 'title'])
```
