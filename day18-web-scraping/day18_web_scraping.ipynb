```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

frames = []

for j in range(1, 6):  # 5 pages for practice
    url = f'https://books.toscrape.com/catalogue/page-{j}.html'
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage, 'lxml')
    
    books = soup.find_all('article', class_='product_pod')
    
    names = []
    prices = []
    ratings = []
    
    for book in books:
        try:
            names.append(book.h3.a['title'])
        except:
            names.append(None)
        try:
            prices.append(book.find('p', class_='price_color').text)
        except:
            prices.append(None)
        try:
            ratings.append(book.p['class'][1])
        except:
            ratings.append(None)
    
    df = pd.DataFrame({
        'name': names,
        'price': prices,
        'rating': ratings
    })
    
    frames.append(df)

final = pd.concat(frames, ignore_index=True)
print(final)
```
