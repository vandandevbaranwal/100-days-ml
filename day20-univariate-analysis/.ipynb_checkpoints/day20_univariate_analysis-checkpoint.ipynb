```python
import pandas as pd
import seaborn as sns
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Categorical Data

## a. Countplot

```python
sns.countplot(df['Survived']) #old method used in video
```

```python
sns.countplot(data=df, x='Survived') #new method
```

```python
sns.countplot(data=df, x='Survived', hue='Survived', palette=['blue', 'yellow'])
```

```python
sns.countplot(df['Embarked']) #old method in video
#df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(data=df, x='Embarked')
```

```python
sns.countplot(data=df, x='Pclass')
```

```python
sns.countplot(data=df, x='Sex')
```

## b. PieChart

```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```
